```python
#import all required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Scatter plot
x=np.random.random(50)*100  # x-array of 50 numbers generated randomly
y=np.random.random(50)*100  # y-array of 50 numbers generated randomly
plt.scatter(x,y)
plt.show() #not required in google colab
```

```python
plt.scatter(x,y,s=70,c='#FF0000',marker="*")  # s -size of points, c-color in hexadecimal
plt.show() # to show the plot
```

```python
#alpha:Helps visualize data better when there are overlapping points.
#Allows you to distinguish density in scatter plots, as darker areas indicate more overlapping points.

x_data=np.random.random(1000)*100  # x-array of 50 numbers generated randomly
y_data=np.random.random(1000)*100  # y-array of 50 numbers generated randomly
plt.scatter(x,y,s=250,c='#FF0000',alpha=0.4)  # s -size of points, c-color in hexadecimal
plt.show() # to show the plot
```

```python
#line chart
x_data=np.arange(50)*1.5 # x-array of 50 numbers generated sequentially
y_data=np.arange(50)*0.5 # y-array of 50 numbers generated sequentially
plt.plot(x_data,y_data)
plt.plot(y_data,x_data)
```

```python
#plot a simple line chart
years=[2000+x for x in range(11)]
sales=[np.random.rand()*1000 for x in range(11)]
plt.plot(years,sales)
plt.xlabel("Years")
plt.xticks(years,rotation=45)
plt.ylabel("Sales")
plt.title("Sales Data over years")
plt.show()
```

```python
#plot 2 line charts with different linestyles and markers
years=[2000+x for x in range(11)]
sales1=[np.random.rand()*1000 for x in range(11)] #sales1 data
sales2=[np.random.rand()*500 for x in range(11)] # sales2 data
#plt.figure() function in Matplotlib is used to create a new figure,
# figsize: Specifies the width and height of the figure in inches.
plt.figure(figsize=(10,5))
#marker: Defines the shape of the marker, e.g., 'o' for circles, 's' for squares
# markersize: Sets the size of the marker
#color: Sets the line color.
#linestyle: Sets the line style, e.g., '-', '--', ':', etc.
#linewidth: Adjusts the thickness of the line.
plt.plot(years,sales1,marker='o',markersize=10,color='red', linestyle='--', linewidth=3)
plt.plot(years,sales2,marker='s',ms=10,c='g',linestyle=':',lw=2)
# xlabel() / ylabel(): Set the labels for the x and y axes.
plt.xlabel("Years",)
plt.ylabel("Sales",{'fontsize':14,'fontweight':'bold','color':'#AB4556'})
#xticks() / yticks(): Customizes the locations and labels of ticks on the axes.
plt.xticks(years,rotation=45)
# title(): Sets the title of the plot.
plt.title("Sales Data over years",{'fontsize':16,'fontweight':'bold','color':'#00AB56'})
# legend(): Adds a legend to the plot, and set its location and appearance
plt.legend(["Sales1","Sales2"],loc="upper right")
plt.show()

```

```python
#Barchart
subjects=["Machine Learning",'AI','Deep Learning',"Python",'Data visualization']
students=[75,100,150,80,70]
plt.bar(subjects,students)
plt.xlabel("Subjects",{'fontsize':16,'fontweight':'bold','color':'#abAB56'})
plt.ylabel("Students",{'fontsize':16,'fontweight':'bold','color':'#AB0B56'})
plt.xticks(rotation=10)
plt.title("Students enrolled in subjects")
plt.show()
```

```python
# Bar chart
subjects=["Machine Learning",'AI','Deep Learning',"Python",'Data visualization']
students=[75,100,150,80,70]
#width: The width of the bars (default is 0.8).
# alpha: The transparency level of the bars (0 is fully transparent, and 1 is fully opaque).
plt.bar(subjects,students,color='maroon',width=0.4,alpha=0.8)
plt.xlabel("Subjects",{'fontsize':16,'fontweight':'bold','color':'#abAB56'})
plt.ylabel("Students",{'fontsize':16,'fontweight':'bold','color':'#AB0B56'})
plt.xticks(rotation=10)
plt.title("Students enrolled in subjects")
# Adding a grid
# axis: Specifies which axis to draw the gridlines on. Options include 'both','x','y'
plt.grid(axis='y',c='g', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Horizontal Barchart
subjects=["Machine Learning",'AI','Deep Learning',"Python",'Data visualization']
students=[75,100,150,80,70]
plt.figure(figsize=(10,5))
plt.barh(subjects,students)
plt.ylabel("Subjects",{'fontsize':16,'fontweight':'bold','color':'#abAB56'})
plt.xlabel("Students",{'fontsize':16,'fontweight':'bold','color':'#AB0B56'})
plt.xticks(rotation=10)
plt.title("Students enrolled in subjects")
plt.grid(axis='x',c='r', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Horizontal Barchart
#plot a simple line chart
years=[2000+x for x in range(11)]
sales=[np.random.rand()*1000 for x in range(11)]
plt.plot(years,sales)
plt.xlabel("Years")
plt.ylabel("Sales")
plt.title("Sales Data over years")
plt.minorticks_on()
plt.grid(which="both",axis='x',c='r', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Generate random data
data = np.random.randn(1000)

# Create a histogram
plt.hist(data)

# Adding titles and labels
plt.title('Normalized Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Density')

# Show the plot
plt.show()
```

```python
# Generate random data
data = np.random.randn(1000)

# Create a histogram with customizations
plt.hist(data, bins=20, alpha=0.6, color='green',edgecolor='black')

# Adding titles and labels
plt.title('Normalized Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Density')

# Show the plot
plt.show()
```

```python
# Pie chart

countries=['India','China','USA','England','Ireland','Germany']
population=[14400,12000,10000,8000,3000,5000]
explodes=[0.2,0,0,0,0.1,0]
#labels: The labels for each slice.
#explode: In this example, [0.2,0,0,0,0.1,0] moves the 'India' away from center (20% of the radius)
# and 'Ireland' slightly (10%) away from the center.
#autopct: Displays the percentage value inside the slices.

plt.pie(population,labels=countries,explode=explodes,autopct='%.2f%%')
plt.title('Population of Countries')

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')
plt.show()
```

```python
#startangle: Rotates the chart so the slices are oriented differently.
#shadow: Adds a shadow effect to the pie chart.
plt.pie(population,labels=countries,explode=explodes,autopct='%.2f%%',startangle=90,shadow=True)
plt.title('Population of Countries with different startangle')
plt.axis('equal')
plt.show()
```

```python
# Box plot
data=np.random.normal(180,20,800)
# Create a boxplot
plt.boxplot(data)

# Add a title and labels
plt.title("Boxplot Example")
plt.ylabel("Values")

# Show the plot
plt.show()
```

```python
# Generate multiple box-plots
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(90, 15, 200)
data3 = np.random.normal(80, 20, 200)

# Create a boxplot for multiple datasets
plt.boxplot([data1, data2, data3], labels=['Data1', 'Data2', 'Data3'])

# Add a title and labels
plt.title("Boxplot for Multiple Datasets")
plt.ylabel("Values")

# Show the plot
plt.show()
```

```python
# Creating multiple figures for multiple charts

countries=['India','China','USA','England','Ireland','Germany']
population=[14400,12000,10000,8000,3000,5000]
explodes=[0.2,0,0,0,0.1,0]
#labels: The labels for each slice.
#explode: In this example, [0.2,0,0,0,0.1,0] moves the 'India' away from center (20% of the radius)
# and 'Ireland' slightly (10%) away from the center.
#autopct: Displays the percentage value inside the slices.
plt.figure(1)
plt.pie(population,labels=countries,explode=explodes,autopct='%.2f%%')
plt.title('Population of Countries')

plt.figure(2)
#startangle: Rotates the chart so the slices are oriented differently.
#shadow: Adds a shadow effect to the pie chart.
plt.pie(population,labels=countries,explode=explodes,autopct='%.2f%%',startangle=90,shadow=True)
plt.title('Population of Countries with different startangle')

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')
plt.show()
```

```python
# Using same figure divided into 2 subplots to show 2 Pie charts

countries=['India','China','USA','England','Ireland','Germany']
population=[14400,12000,10000,8000,3000,5000]
explodes=[0.2,0,0,0,0.1,0]
#labels: The labels for each slice.
#explode: In this example, [0.2,0,0,0,0.1,0] moves the 'India' away from center (20% of the radius)
# and 'Ireland' slightly (10%) away from the center.
#autopct: Displays the percentage value inside the slices.
plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.pie(population,labels=countries,explode=explodes,autopct='%.2f%%')
plt.title('Population of Countries')

plt.subplot(1,2,2)
#startangle: Rotates the chart so the slices are oriented differently.
#shadow: Adds a shadow effect to the pie chart.
plt.pie(population,labels=countries,explode=explodes,autopct='%.2f%%',startangle=90,shadow=True)
plt.title('Population of Countries with different startangle')

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')
plt.show()
```

```python
#Creating muliple subplots

fig,axes=plt.subplots(2,2,figsize=(14,10))
# 1st subplot
axes[0,0].plot(years,sales1)
axes[0,0].set_title("Sales1 over years")
axes[0,0].set_xlabel("Years")
axes[0,0].set_ylabel("Sales")

# 2nd subplot
axes[0,1].plot(years,sales2)
axes[0,1].set_title("Sales2 over years")
axes[0,1].set_xlabel("Years")
axes[0,1].set_ylabel("Sales")


# 3rd subplot
axes[1,0].bar(subjects,students)
axes[1,0].set_title("Students enrolled in subjects")
axes[1,0].set_xlabel("Subjects")
axes[1,0].set_ylabel("Students")

#4th subplot
axes[1,1].pie(population,labels=countries,autopct='%.2f%%')
axes[1,1].set_title('Population of Countries')
fig.suptitle("Multiple Plots")
plt.savefig("multiple_plots.png") #saving the figure
plt.show()
```

```python
#generating 3D Scatter graph

ax=plt.axes(projection='3d')

x=np.random.random(200)
y=np.random.random(200)
z=np.random.random(200)
# #we cant rotate the 3d graph in colab use ax.view_init() to set a specific viewing angle
ax.view_init(elev=30,azim=20)
ax.scatter(x,y,z,c='r')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

#ax.scatter(x,y,z,c='r')


```

```python
#generating 3D Scatter graph

ax=plt.axes(projection='3d')

x=np.arange(0,100,0.2)
y=np.cos(x)
z=np.sin(x)
ax.plot(x,y,z,c='g')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

```python
#plot the 3D surface
ax=plt.axes(projection='3d')

x=np.arange(-7,7,0.2)
y=np.arange(-7,7,0.2)
X,Y=np.meshgrid(x,y)
Z=np.sin(X)+np.cos(Y)
# Plot the surface
ax.plot_surface(X,Y,Z,cmap='viridis')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
```

```python
# bar chart with animation  -part-1
'''Note: In Google Colab, the standard plt.pause() method used for dynamic plot updating does not work as expected
because of how Colab handles rendering. Instead, you can update plots using IPython.display to clear and re-draw the plot within the loop.'''
import random
from IPython.display import clear_output


# Initialize counts for heads and tails
head_tails = [0, 0]
# Loop to update the chart dynamically
for _ in range(50):# You can increase this number to simulate more flips
    # Increment heads or tails randomly
    head_tails[random.randint(0, 1)] += 1
    # Create a new bar chart
    plt.bar(['Head', 'Tail'],head_tails,color=['red','blue'])

     # Redraw the chart with a short pause to create the animation effect
    plt.pause(0.001)

plt.show()

```

```python
# Bar chart with animation -part-2
import random
from IPython.display import clear_output

# Initialize counts for heads and tails
head_tails = [0, 0]
# Loop to update the chart dynamically
for _ in range(100):# You can increase this number to simulate more flips
    # Increment heads or tails randomly
    head_tails[random.randint(0, 1)] += 1
    # Clear the previous plot
    clear_output(wait=True)
    # Create a new bar chart
    plt.bar(['Head', 'Tail'],head_tails,color=['red','blue'])

     # Redraw the chart with a short pause to create the animation effect
    plt.pause(0.001)

plt.show()
```
