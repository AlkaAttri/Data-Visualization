```python
# import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#read the csv file or dataset
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Visualisation Project/VideoGamesSales.csv')
df
```

```python
#size of dataset
df.shape
```

```python
#check if any row is duplicated
df.duplicated().any()
```

```python
# showing duplicated rows
df_duplicated_rows=df[df.duplicated()]
print(df_duplicated_rows.shape)
df_duplicated_rows
```

```python
#drop duplicated rows
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.info()  # one - way to check for null values
```

```python
#other way to check for NaN or null values
filter_na=df.isna().any()
print(filter_na)
filter_na=filter_na.loc[filter_na]
filter_na
```

```python
#access the values at specific row and na values
df_na=df[df['Publisher'].isna()]
print(df_na.shape)
df_na
```

```python
df['Publisher'].unique()
```

```python
#df['Publisher'].fillna('OTHER_PUBLISHER',inplace=True)  # The behavior will change in pandas 3.0.
df['Publisher']=df['Publisher'].fillna('OTHER_PUBLISHER')
df['Publisher'].unique()
```

```python
df_region_na=df[df['Region'].isna()]
print(df_region_na.shape)
df_region_na
```

```python
df['Region'].unique()
```

```python
# replace NaN value with 'OTHER' in Region column
df['Region']=df['Region'].fillna('OTHER')
df['Region'].unique()
```

```python
df['NA_Sales'].unique()
```

```python
#Replace dollar $ sign with  ' ' space
df['NA_Sales']=df['NA_Sales'].replace('[$]',' ',regex=True)
df['NA_Sales']
```

```python
df.columns
```

```python
df['Country'].unique()
```

```python
df['Country']=df['Country'].replace("United States","USA")
df['Country'].unique()
```

```python
df[df['Country']=="USA"]
```

```python
# Renaming the coulumns in proper format
df.rename(columns={'NA_Sales':'National Sales', 'Global_Sales':'Global Sales','NA_Profit':"National Profit",'Global_Profit':"Global Profit"},inplace=True)
df.columns
```

```python
df.head(3)
```

```python
#converting the datatype of National Sales column to float
df['National Sales']=df['National Sales'].astype(float)

df.dtypes
```

```python
# Total National Sales Region and country wise
df_group=df.groupby(['Region','Country'])['National Sales'].sum()
df_group
```

```python
#reset index of above Series
df_group=df_group.reset_index()
df_group
```

```python
# sort the values in decreasing order by Natonal Sales
df_group=df_group.sort_values('National Sales',ascending=False)
df_group
```

```python
#visualisation using bar graph

import warnings # To suppress some warnings

# Suppress the specific FutureWarning
warnings.filterwarnings("ignore", category=FutureWarning, module="seaborn")

ax=sns.barplot(y='National Sales',x='Region',data=df_group,hue='Country')
# Adding text on the bars
for p in ax.patches:  # ax.patches is a list of the bars
    height = p.get_height()  # Get the height of each bar
    ax.text(
        p.get_x() + p.get_width() / 2,  # x position: center of the bar
        height,  # y position: height of the bar
        round(height, 2),  # Text to display
        ha="center",  # Center horizontally
        va="bottom"  # Place text above the bar
    )

plt.title('National Sales Region and Country wise',{'fontsize':16,'fontweight':'bold','color':'green'})
plt.ylabel('National Sales',{'fontsize':14,'fontweight':'bold','color':'#0000FF'})
plt.xlabel('Region',{'fontsize':14,'fontweight':'bold','color':'#AB4556'})
plt.show()
```

```python
#country wise national sales and global sales
df_ns_gs=df.groupby(['Country'])[['National Sales','Global Sales']].sum()
df_ns_gs
```

```python
df_ns_gs.reset_index(inplace=True)
df_ns_gs
```

```python
#pie chart visualisation
plt.subplot(1,2,1)
plt.pie(df_ns_gs['National Sales'],labels=df_ns_gs['Country'],autopct='%.2f%%',startangle=90,colors="gy")
plt.title('National Sales')

plt.subplot(1,2,2)
plt.pie(df_ns_gs['Global Sales'],labels=df_ns_gs['Country'],autopct='%.2f%%',startangle=270)
plt.title('Global Sales')
```

```python
#find year-wise Global Profit and National Profit
df_gf_nf_year=df.groupby(['Year'])[['Global Profit','National Profit']].sum()
df_gf_nf_year
```

```python
# create a dataframe
df_gf_nf_year.reset_index(inplace=True)
df_gf_nf_year
```

```python
#line Chart
plt.plot(df_gf_nf_year['Year'],df_gf_nf_year['Global Profit'],label='Global Profit',marker="s")
plt.plot(df_gf_nf_year['Year'],df_gf_nf_year['National Profit'],label='National Profit',marker='o')
plt.legend()
plt.xlabel('Year',{'fontsize':14,'fontweight':'bold','color':'#AB4556'})
plt.ylabel('Profit',{'fontsize':14,'fontweight':'bold','color':'#AB00FF'})
plt.show()
```

```python
#find Genre
df['Genre'].unique()
```

```python
#show the genre -wise National Sales
df_genre=df.groupby(['Genre'])[['National Sales','Global Sales']].sum().reset_index().sort_values('National Sales',ascending=False)
df_genre
```

```python
#Bar-Chart showing Genre-wise National Sales
ax=sns.barplot(y='Genre',x='National Sales',data=df_genre,hue="Genre")
# Adding text on the horizontal bars
for p in ax.patches:
    width = p.get_width()  # Get the width of each bar (i.e., the value)
    ax.text(
        width,  # x position: end of the bar
        p.get_y() + p.get_height() / 2,  # y position: center of the bar
        round(width, 2),  # Text to display (rounded value)
        ha='left',  # Align text to the left of the bar's end
        va='center',  # Center text vertically along the bar
        color='brown'
    )
plt.title('Genre -Wise National Sales',{'fontsize':14,'fontweight':'bold'})
plt.xlabel('National Sales')
plt.ylabel('Genre')
plt.show()
```

```python
ax=sns.barplot(y='Genre',x='Global Sales',data=df_genre.sort_values('Global Sales',ascending=False))
# Adding text on the horizontal bars
for p in ax.patches:
    width = p.get_width()  # Get the width of each bar (i.e., the value)
    ax.text(
        width,  # x position: end of the bar
        p.get_y() + p.get_height() / 2,  # y position: center of the bar
        round(width, 2),  # Text to display (rounded value)
        ha='left',  # Align text to the left of the bar's end
        va='center',  # Center text vertically along the bar
        color='brown'
    )

plt.title('Genre -Wise Global Sales',{'fontsize':14,'fontweight':'bold'})
plt.xlabel('Global Sales')
plt.ylabel('Genre')
plt.show()
```

```python
sns.barplot(y="Genre", x="National Sales", hue="Global Sales", data=df_genre)
```

```python
#stacked bar chart
bar1=plt.barh(df_genre['Genre'],df_genre['Global Sales'],label='Global Sales',color='brown')
bar2=plt.barh(df_genre['Genre'],df_genre['National Sales'],label='National Sales',color='r')
for bar in bar1:
    width = bar.get_width()  # Get width of the bar (i.e., sales value)
    plt.text(
        width,  # x position: width of the bar
        bar.get_y() + bar.get_height()/2,  # y position: center of the bar
        round(width, 2),  # Text to display (rounded sales value)
        ha='left',  # Align text to the left of the bar's end
        va='center',  # Center text vertically along the bar
        color='maroon' #set the color
    )
for bar in bar2:
    width = bar.get_width()  # Get width of the bar (i.e., sales value)
    plt.text(
        width,  # x position: width of the bar
        bar.get_y() + bar.get_height()/2,  # y position: center of the bar
        round(width, 2),  # Text to display (rounded sales value)
        ha='center',  # Align text to the center of the bar's end
        va='center' # Center text vertically along the bar

    )
plt.legend()
plt.ylabel('Genre',{'fontsize':14,'fontweight':'bold','color':'#AB4556'})
plt.xlabel('Sales',{'fontsize':14,'fontweight':'bold','color':'#AB00FF'})

plt.show()
```
