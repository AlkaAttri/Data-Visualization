Some **Math module** functions

```python
from math import ceil, floor, trunc,factorial

x = 1.4
y = 2.6

print(floor(x), floor(y))
print(floor(-x), floor(-y))
print(ceil(x), ceil(y))
print(ceil(-x), ceil(-y))
print(trunc(x), trunc(y))
print(trunc(-x), trunc(-y))
print(factorial(11))
print(math.sqrt(6),math.isqrt(6))
print(math.fabs(-3.7)) #math module function only for float values & dont support complex no.
print(abs(-7.8)) # built in function for int or float and support complex no also
```

**The random function**

 produces a float number x coming from the range (0.0, 1.0) - in other words: (0.0 <= x < 1.0).

```python
from random import random
for i in range(5):
  print(random())
```

**The seed function**

The seed() function is able to directly set the generator's seed.

seed() - sets the seed with the current time

seed(int_value) - sets the seed with the integer value int_value.

```python
from random import seed
seed()
for i in range(3):
  print(random())
```

```python
seed(6)
for i in range(2):
  print(random())
```

 **The randrange function**

randrange(end)

randrange(beg, end)

randrange(beg, end, step)

**note:** right-sided exclusion

```python
from random import randrange
for i in range(10):
  print(randrange(11),end=' ')
```

```python
for i in range (11):
  print(randrange(1,11),end=' ')
```

```python
for i in range(11):
  print(randrange(0,101,10),end=' ')
```

 **The randint function**

 randint(left, right)   
** note:** right included

```python
from random import randint
for i in range(10):
  print(randint(10,20),end=' ')
```

**The choice function**

chooses a "random" element from the input sequence and returns it

```python
from random import randint,choice
my_list= [randint(1,10) for i in range(11)] #list comprehension
print("My list = ", my_list)
print("Selected number : ",choice(my_list))
```

**The sample function**

 returns a new list containing k  elements chosen randomly from the given population.

 sample() selects k unique indices



```python
from random import sample
my_list= [randint(1,10) for i in range(11)]
print("My list = ", my_list)
print("Selected sublist : ",sample(my_list,4))
```

```python
sub_list=[i for i in range(10)]
print("Sub list = ", sub_list)
print("Selected sublist : ",sample(sub_list,6))
```

**The platform module**
The platform module lets you access the underlying platform's data, i.e., hardware, operating system, and interpreter version information.
**The platform function**

platform(aliased=False, terse=False)

aliased=True → may show alternative names for OS layers

terse=True → tries to show a shorter version of the result


```python
from platform import platform
print(platform())
print(platform(1))
print(platform(0, 1))
```

**The machine function**

returns the generic architecture of machine

e.g x86_64 → 64‑bit Intel/AMD processors

```python
from platform import machine
print(machine())
```

**The processor function**- returns real processor name or CPU model

```python
from platform import processor
print(processor())
```

**The system function**- returns os

```python
from platform import system,version
print(system()) #returns os(operating system)
version()  #returns version of os
```

**python_implementation, python_version,python_version_tuple funtions**

```python
from platform import python_implementation, python_version,python_version_tuple
print(python_implementation()) #tells you which Python interpreter is running your cod
print(python_version()) # returns python version
print(python_version_tuple()) #returns Python version as a 3‑element tuple of strings: (major, minor, micro).
```
